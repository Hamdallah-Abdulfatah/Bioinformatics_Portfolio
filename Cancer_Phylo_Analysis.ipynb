```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Collect Data (Simulated Dataset)
cancer_mutations = {
    "Gene": ["TP53", "KRAS", "BRAF", "EGFR", "PIK3CA"],
    "Lung Cancer": [60, 35, 10, 70, 30],
    "Breast Cancer": [50, 10, 20, 15, 80],
    "Colorectal Cancer": [40, 65, 55, 25, 45],
    "Pancreatic Cancer": [50, 80, 35, 20, 25]
    }
```

```python
# Process Data
mutation_df = pd.DataFrame(cancer_mutations)
mutation_df.set_index("Gene", inplace=True)
```

```python
# Visualize Data - Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(mutation_df, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Common Cancer Mutations and Their Frequencies")
plt.xlabel("Cancer Type")
plt.ylabel("Mutated Gene")
plt.show()
```

```python
# Interpretation of Results
print("\nInterpretation:")
print("- TP53 is highly mutated across all cancer types, suggesting it is a major tumor suppressor gene.")
print("- KRAS mutations are very frequent in pancreatic and colorectal cancers.")
print("- EGFR mutations are dominant in lung cancer, indicating its role in targeted therapy.")
```
